```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
epoch_10 = pd.read_csv('epoch10', sep=',', engine='python')
epoch_25 = pd.read_csv('epoch25', sep=',', engine='python')
epoch_50 = pd.read_csv('epoch50', sep=',', engine='python')
epoch_100 = pd.read_csv('epoch100', sep=',', engine='python')
epoch_200 = pd.read_csv('epoch200', sep=',', engine='python')
epoch_500 = pd.read_csv('epoch500', sep=',', engine='python')
epoch_1000 = pd.read_csv('epoch1000', sep=',', engine='python')
epoch_1500 = pd.read_csv('epoch1500', sep=',', engine='python')
epoch_2000 = pd.read_csv('epoch2000', sep=',', engine='python')

epoch_labels = [10, 25, 50, 100, 200, 500, 1000, 1500, 2000]
epoch_mean_data = [epoch_10['val_loss'].mean(), epoch_25['val_loss'].mean(), epoch_50['val_loss'].mean(), epoch_100['val_loss'].mean(), epoch_200['val_loss'].mean(), epoch_500['val_loss'].mean(), epoch_1000['val_loss'].mean(), epoch_1500['val_loss'].mean(), epoch_2000['val_loss'].mean()]
```

```python
batch_1 = pd.read_csv('batch1', sep=',', engine='python')
batch_4 = pd.read_csv('batch4', sep=',', engine='python')
batch_8 = pd.read_csv('batch8', sep=',', engine='python')
batch_16 = pd.read_csv('batch16', sep=',', engine='python')
batch_32 = pd.read_csv('batch32', sep=',', engine='python')
batch_64 = pd.read_csv('batch64', sep=',', engine='python')
batch_128 = pd.read_csv('batch128', sep=',', engine='python')

batch_labels = [1, 4, 8, 16, 32, 64, 128]
batch_mean_data = [batch_1['val_loss'].mean(), batch_4['val_loss'].mean(), batch_8['val_loss'].mean(), batch_16['val_loss'].mean(), batch_32['val_loss'].mean(), batch_64['val_loss'].mean(), batch_128['val_loss'].mean()]
```

```python
plt.xlabel('Number of Epochs (Training Iterations)')
plt.ylabel('Average Loss')
plt.title('Epoch Analysis')
plt.plot(epoch_labels, epoch_mean_data, color = 'red')
```

```python
plt.xlabel('Batch Sizes')
plt.ylabel('Average Loss')
plt.title('Batch Size Analysis')
plt.plot(batch_labels, batch_mean_data, color = 'red')
```
